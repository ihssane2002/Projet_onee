```python
import os
import rarfile
from PIL import Image
import matplotlib.pyplot as plt

# Chemins
input_folder = "/content/drive/MyDrive/Incidents_Images/images.rar"  # Fichier .rar contenant les images
extracted_folder = "/content/drive/MyDrive/Incidents_Images/extracted_images"  # Dossier d'extraction des images

# Extraire le fichier .rar
with rarfile.RarFile(input_folder) as rf:
    rf.extractall(extracted_folder)  # Extraire dans le dossier spécifié

print(f"Fichiers extraits dans {extracted_folder}.")

# Chemin du sous-dossier contenant les images
images_folder = os.path.join(extracted_folder, "images")

# Vérifier que les images ont bien été extraites
extracted_files = os.listdir(images_folder)
print(f"Nombre de fichiers extraits : {len(extracted_files)}")
print("Exemples de fichiers extraits :", extracted_files[:5])  # Montre les 5 premiers fichiers

# Afficher 5 images pour vérification
plt.figure(figsize=(15, 10))
for i, img_name in enumerate(extracted_files[:5]):
    img_path = os.path.join(images_folder, img_name)
    try:
        if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            img = Image.open(img_path)
            plt.subplot(1, 5, i + 1)
            plt.imshow(img)
            plt.title(img_name)
            plt.axis('off')
        else:
            print(f"Fichier non image ignoré : {img_name}")
    except Exception as e:
        print(f"Erreur avec l’image {img_name} : {e}")

plt.show()
```

```python
!pip install git+https://github.com/openai/CLIP.git
import torch
import clip
import csv
from tqdm import tqdm
from google.colab import files

# Charger le modèle CLIP
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device)

# Liste des classes
incident_types = [
    "Compteur brûlé", "Poteau tombé", "Câble endommagé",
    "Oiseau sur les lignes", "Isolateur cassé", "Arbre sur les lignes"
]
text_inputs = torch.cat([clip.tokenize(f"Une image de {c}") for c in incident_types]).to(device)

# Chemin du fichier CSV
csv_output_path = "/content/classified_incidents.csv"

# Initialiser le fichier CSV et écrire les en-têtes
with open(csv_output_path, mode="w", newline="") as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(["Image_Name", "Predicted_Class"])  # En-têtes du CSV

    # Parcourir les fichiers extraits
    for img_name in tqdm(extracted_files):
        img_path = os.path.join(images_folder, img_name)
        try:
            # Vérifier si l'image est un fichier valide
            if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                # Ouvrir et prétraiter l'image
                image = Image.open(img_path)
                image = preprocess(image).unsqueeze(0).to(device)

                with torch.no_grad():
                    logits_per_image, _ = model(image, text_inputs)
                    probs = logits_per_image.softmax(dim=-1).cpu().numpy()

                # Prédire la classe de l'image
                predicted_class = incident_types[probs.argmax()]

                # Écrire le nom de l'image et la classe prédite dans le CSV
                writer.writerow([img_name, predicted_class])
            else:
                print(f"Fichier non image ignoré : {img_name}")

        except Exception as e:
            print(f"Erreur avec l’image {img_name} : {e}")

print(f"Classification terminée ! 🚀 Le fichier CSV a été sauvegardé ici : {csv_output_path}")

# Télécharger le fichier CSV sur votre ordinateur
files.download(csv_output_path)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import pandas as pd

# Charger le fichier CSV
df = pd.read_csv("/content/balanced_classified_incidents.csv")

# Séparer les données en ensembles d'entraînement et de test
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Utiliser ImageDataGenerator pour charger et prétraiter les images
datagen = ImageDataGenerator(rescale=1./255)

train_generator = datagen.flow_from_dataframe(
    dataframe=train_df,
    directory="/content/drive/MyDrive/Incidents_Images/extracted_images/images",
    x_col="Image_Name",
    y_col="Predicted_Class",
    target_size=(224, 224),
    batch_size=32,
    class_mode="categorical"
)

test_generator = datagen.flow_from_dataframe(
    dataframe=test_df,
    directory="/content/drive/MyDrive/Incidents_Images/extracted_images/images",
    x_col="Image_Name",
    y_col="Predicted_Class",
    target_size=(224, 224),
    batch_size=32,
    class_mode="categorical"
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Créer un modèle CNN simple
model = Sequential([
    Conv2D(32, (3, 3), activation="relu", input_shape=(224, 224, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation="relu"),
    Dense(len(train_generator.class_indices), activation="softmax")  # Nombre de classes
])

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

# Ajouter EarlyStopping pour éviter l'overfitting
early_stopping = EarlyStopping(
    monitor='val_loss',  # Surveiller la perte de validation
    patience=5,  # Nombre d'époques sans amélioration avant l'arrêt
    restore_best_weights=True  # Restaurer les meilleurs poids du modèle
)

# Entraîner le modèle avec le callback EarlyStopping
model.fit(
    train_generator,
    epochs=20,
    validation_data=test_generator,
    callbacks=[early_stopping]  # Ajouter le callback
)
```

```python
loss, accuracy = model.evaluate(test_generator)
print(f"Précision du modèle : {accuracy * 100:.2f}%")
```
